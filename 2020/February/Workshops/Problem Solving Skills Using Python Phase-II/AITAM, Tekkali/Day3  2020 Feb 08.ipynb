```python
roll = '18A51A04'
rollNumbers = []
for num in range(1,101):
    if num < 10:
        s = roll + str(0) + str(num)
    else:
        s = roll + str(num)
    rollNumbers.append(s)
print(rollNumbers)
```

```python
import numpy as np
python = [np.random.randint(0,100) for i in range(1,101)]
C = [np.random.randint(0,100) for i in range(1,101)]
pandas = [np.random.randint(0,100) for i in range(1,101)]
numpy = [np.random.randint(0,100) for i in range(1,101)]
numpy
```

## username@domainName.extension

rollNumber@apssdc.in

```python
email = [roll + '@apssdc.in' for roll in rollNumbers]
email
```

```python
data = {'Roll Numbers':rollNumbers,'Email': email,'C':C,
 'Python':python,'Pandas':pandas,'Numpy':numpy}
print(data)
```

```python
sdata = pd.DataFrame(data)
sdata
```

```python
sdata['TotalMarks'] = sdata['C'] + sdata['Python'] + sdata['Numpy'] + sdata['Pandas']
sdata
```

```python
sdata['Percentage'] = (sdata['TotalMarks'] * 100) / 400
sdata
```

```python
sdata['Status'] = ['Pass' if data > 40 else 'Fail' for data in sdata['Percentage']]
sdata
```

```python
sdata.to_excel('sdata.xlsx',sheet_name='Class1')
```

```python
sdata.to_csv('sdata.csv')
```
